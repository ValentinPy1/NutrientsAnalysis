```python

# Create a correlation matrix
corr_matrix = df[num_cols].corr()

# Create a heatmap without numbers
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Create a KMeans model with 10 clusters, k-means++ initialization, and 20 initializations
kmeans = KMeans(n_clusters=30, init='k-means++', n_init=20)

# Fit the model to the data and predict the cluster labels
df['KMeans'] = kmeans.fit_predict(df[num_cols])

# Print the number of data points in each cluster
print(df['KMeans'].value_counts())
```

```python
from sklearn_extra.cluster import KMedoids

# Create a KMedoids model with 10 clusters
kmedoids = KMedoids(n_clusters=30, method='pam')

# Fit the model to the data and predict the cluster labels
df['KMenoids'] = kmedoids.fit_predict(df[num_cols])

# Print the number of data points in each cluster
print(df['KMenoids'].value_counts())
```

```python
import umap

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Create a UMAP
reducer = umap.UMAP()
embedding = reducer.fit_transform(df[num_cols])
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(embedding[:, 0], embedding[:, 1], c=df['KMeans'])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection with KMeans clustering', fontsize=12)
plt.colorbar(label='cluster')

# Plot the UMAP with colors based on the KMedoids clustering
plt.subplot(1, 2, 2)
plt.scatter(embedding[:, 0], embedding[:, 1], c=df['KMenoids'])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection with KMedoids clustering', fontsize=12)
plt.colorbar(label='cluster')

plt.tight_layout()
plt.show()
```
