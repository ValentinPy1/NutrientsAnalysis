```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '../Data/DRI/VitaminsRDA.txt'
df = pd.read_csv(file)
df
```

```python
def convert_age(age):
    if 'mo' in age:
        return int(age.split('-')[0]) / 12
    if '>' in age:
        return 70
    return int(age.split('-')[0])

df['Age'] = df['Age'].apply(convert_age)
```
